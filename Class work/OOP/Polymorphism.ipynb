```python
class Product:
  Cmp = "AMAZON"
  a = 52
  a = 50
  def show(self):
    print("1")
  def show(self,a=20):
    print("2")

  def show(self,x,a=100):
    print("3")

class Purchase(Product):
  newCmp = "Myntra"
  def show(self,a=10):
    print(a)

obj = Purchase()
obj.show(99)
```
