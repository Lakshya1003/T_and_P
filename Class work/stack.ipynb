Write a program to perform push operation in Stack.

```python
def createStack():
    radar = list()
    return radar

def push(radar, n):
    radar.append(n)
    print("pushed item: " + n)

def show(radar):
    print("The stack elements are:")
    for i in radar:
        print(i)
```

```python
stack = createStack()
push(stack, str(10))
push(stack, str(20))
push(stack, str(30))
push(stack, str(40))
show(stack)
```
