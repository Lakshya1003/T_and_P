Question : Write a program to find the passward ;<br>
input1 : 12 <br>
input1 : 1212<br>
input1 : 567<br>
input1 : 233<br>
input1 : 8877<br>

output : Sum of stable numbers - sum of unstable numbers <br>

stable number = number whos all unique elements have same frequency

```python
inputs = [12, 1212, 567, 233, 8877]

stable_sum = 0
unstable_sum = 0

for num in inputs:
    s = str(num)
    freq = {digit: s.count(digit) for digit in set(s)}

    # check if all frequencies are equal
    if len(set(freq.values())) == 1:
        stable_sum += num
    else:
        unstable_sum += num

print("Output =", stable_sum - unstable_sum)
```
